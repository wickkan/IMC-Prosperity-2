```python
import pandas as pd

prices_round_4_day_1 = pd.read_csv('prices_round_4_day_1.csv', delimiter=';')
prices_round_4_day_2 = pd.read_csv('prices_round_4_day_2.csv', delimiter=';')
prices_round_4_day_3 = pd.read_csv('prices_round_4_day_3.csv', delimiter=';')
trades_round_4_day_1 = pd.read_csv('trades_round_4_day_1_nn.csv', delimiter=';')
trades_round_4_day_2 = pd.read_csv('trades_round_4_day_2_nn.csv', delimiter=';')
trades_round_4_day_3 = pd.read_csv('trades_round_4_day_3_nn.csv', delimiter=';')
```

```python
# Combining the price data frames for easier analysis
prices = pd.concat([prices_round_4_day_1, prices_round_4_day_2, prices_round_4_day_3])

# Basic statistics for COCONUT prices
print('Basic statistics for COCONUT prices')
print(prices[prices['product'] == 'COCONUT'].describe())

# Basic statistics for COCONUT_COUPON prices
print('Basic statistics for COCONUT_COUPON prices')
print(prices[prices['product'] == 'COCONUT_COUPON'].describe())
```

```python
# Combining the trade data frames for easier analysis
trades = pd.concat([trades_round_4_day_1, trades_round_4_day_2, trades_round_4_day_3])

# Basic statistics for COCONUT_COUPON trades
print('Basic statistics for COCONUT_COUPON trades')
print(trades[trades['symbol'] == 'COCONUT_COUPON'].describe())
```

```python
import matplotlib.pyplot as plt

# Plotting mid-price trends for COCONUT
prices_coconut = prices[prices['product'] == 'COCONUT']
plt.figure(figsize=(12, 6))
plt.plot(prices_coconut['timestamp'], prices_coconut['mid_price'], label='COCONUT Mid Price')
plt.title('COCONUT Mid Price Trends')
plt.xlabel('Timestamp')
plt.ylabel('Mid Price')
plt.legend()
plt.show()

# Plotting mid-price trends for COCONUT_COUPON
prices_coconut_coupon = prices[prices['product'] == 'COCONUT_COUPON']
plt.figure(figsize=(12, 6))
plt.plot(prices_coconut_coupon['timestamp'], prices_coconut_coupon['mid_price'], label='COCONUT_COUPON Mid Price', color='red')
plt.title('COCONUT_COUPON Mid Price Trends')
plt.xlabel('Timestamp')
plt.ylabel('Mid Price')
plt.legend()
plt.show()
```

```python
# Calculating daily returns
prices['daily_return'] = prices.groupby('product')['mid_price'].pct_change()

# Calculating Exponential Moving Average of daily returns as a measure of volatility
prices['volatility'] = prices.groupby('product')['daily_return'].transform(lambda x: x.ewm(span=50).std())

# Plotting the EMA of volatility for both products
plt.figure(figsize=(12, 6))
for product in ['COCONUT', 'COCONUT_COUPON']:
    subset = prices[prices['product'] == product]
    plt.plot(subset['timestamp'], subset['volatility'], label=f'{product} Volatility')
plt.title('Volatility Trends Using EMA of Daily Returns')
plt.xlabel('Timestamp')
plt.ylabel('Volatility')
plt.legend()
plt.show()
```

```python
# Summarizing trade volumes by day
trade_volumes_by_day = trades.groupby('timestamp')['quantity'].sum()
print(trade_volumes_by_day)
```

```python
# Identifying the times when prices are at their lowest or highest
prices['is_low'] = (prices['mid_price'] <= prices.groupby('product')['mid_price'].transform('min'))
prices['is_high'] = (prices['mid_price'] >= prices.groupby('product')['mid_price'].transform('max'))

# Counting occurrences of lows and highs
low_high_counts = prices.groupby(['product', 'is_low', 'is_high']).size().unstack(fill_value=0)
print(low_high_counts)
```

```python
# Filtering data to include only the rows where extremes are reached
extreme_prices = prices[(prices['is_low'] == True) | (prices['is_high'] == True)]

# Investigating the conditions when extremes are hit
extreme_conditions = extreme_prices[['timestamp', 'product', 'mid_price', 'bid_price_1', 'ask_price_1', 'is_low', 'is_high']]
print(extreme_conditions)
```
